```python
#using lowercase 'women' instead of 'Women' finds more docs, since the term isn't always capitalized in the df
#as in subject headings relating to laws about "Married women" as opposed to "Women--Legal status" or docs with "women" in the title
Women_allcols = df.apply(lambda x: x.str.contains('women', na=False), axis=1).any(axis=1)
```

```python
#sorting by dates to get sense of timeline
#this set of 14 docs addresses immigration laws concerning women from the 1890s-1930s
#https://builtin.com/data-science/pandas-show-all-columns
df[Women_allcols].sort_values('Dates')
pd.set_option('max_colwidth', None)
```

```python
df[Women_allcols]
```

```python
#Bar chart providing most frequent date ranges for docs about Chinese Exclusion
#Shows most date from the earlier periods (1890-1900 & 1900-1910) when Chinese immigration polices were debated, studied and written
df[Chinese_allcols].value_counts('Date Range').plot(kind='bar', title='Documents Timeline: Chinese Exclusion', ylabel='# of Documents', color='purple')
```

```python
#whereas for immigration policies concerning women, most docs date to later decades
#during the 1920s and 1930s, when the issue of citizenship status for American women married to "aliens" was revisited
#as well as laws about the wives of American citizens of Asian descent
df[Women_allcols].value_counts('Date Range').plot(kind='bar', title='Documents Timeline: Women & Immigration', ylabel='# of Documents', color='Crimson')
plt.savefig('Women_docs.png', bbox_inches='tight')
```
