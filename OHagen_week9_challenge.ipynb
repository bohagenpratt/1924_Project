```python
#get data into lists using append
ids = []
title = []
link = []
description = []
subjects = []
creator = []
contributor = []
date = []
location = []
for i in immig_docs:
  id = i.get('id')
  titl = i['sourceResource'].get('title')
  urls = i.get('isShownAt')
  descs = i['sourceResource'].get('description')
  creators = i['sourceResource'].get('creator')
  contribs = i['sourceResource'].get('contributor')
  subjs = i['sourceResource'].get('subject')
  locs = i['sourceResource'].get('spatial')
  dat = i['sourceResource']['date'][0].get('displayDate')
  ids.append(id)
  title.append(titl)
  link.append(urls)
  description.append(descs)
  date.append(dat)
  location.append(locs)
  subjects.append(subjs)
  creator.append(creators)
  contributor.append(contribs)
```
